traemos un modelo(embeddings)  pre entrenado, llamado word2vec, para que nos del los vectores de 3 palabras. y nosotros luego con numpy , aplicamos operaciones de algebra, para compara similitud

```python
import numpy as np
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity

# Supongamos que tienes un modelo Word2Vec preentrenado llamado 'modelo_word2vec'
# Cargar el modelo (reemplaza 'ruta_del_modelo' con la ruta correcta)
modelo_word2vec = Word2Vec.load('ruta_del_modelo')

# Obtener vectores para las palabras 'camioneta', 'auto' y 'pera'
vector_camioneta = modelo_word2vec.wv['camioneta']
vector_auto = modelo_word2vec.wv['auto']
vector_pera = modelo_word2vec.wv['pera']

# Calcular similitudes coseno
similitud_auto_camioneta = cosine_similarity([vector_auto], [vector_camioneta])[0][0]
similitud_auto_pera = cosine_similarity([vector_auto], [vector_pera])[0][0]
similitud_camioneta_pera = cosine_similarity([vector_camioneta], [vector_pera])[0][0]

# Imprimir similitudes
print(f"Similitud coseno entre 'auto' y 'camioneta': {similitud_auto_camioneta}")
print(f"Similitud coseno entre 'auto' y 'pera': {similitud_auto_pera}")
print(f"Similitud coseno entre 'camioneta' y 'pera': {similitud_camioneta_pera}")
```
