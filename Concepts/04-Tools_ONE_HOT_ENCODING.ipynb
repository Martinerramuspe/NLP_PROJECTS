### One-Hot Encoding
Hace referencia al ejercicio que realizamos con la técnica de Bolsa de Palabras (BoW).

```python
# Importamos las bibliotecas necesarias
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Supongamos que tienes un DataFrame 'df' con una columna 'comentarios'
# Puedes cargar tus datos reales aquí
df = pd.DataFrame({'comentarios': ["Este es un comentario.", "Otro comentario interesante.", "Un comentario más."]})
```

```python
df.head()
```

```python
df.shape
```

```python
# Paso 1: Extracción de Comentarios
comentarios = df['comentarios']
```

```python
comentarios.shape
```

```python
comentarios
```

```python
# Paso 2: Vectorización de Palabras o llamando en ingles "Bag of Words"
vectorizador = CountVectorizer()
matriz_bolsa_de_palabras = vectorizador.fit_transform(comentarios)
```

```python
print(matriz_bolsa_de_palabras.toarray())
```

```python
print(matriz_bolsa_de_palabras)
```

```python
# Obtén las palabras únicas como características y crea un DataFrame
palabras_unicas = vectorizador.get_feature_names_out()
matriz_palabras = pd.DataFrame(matriz_bolsa_de_palabras.toarray(), columns=palabras_unicas)
```

```python
print(palabras_unicas)
```

```python
print(matriz_palabras)
```
