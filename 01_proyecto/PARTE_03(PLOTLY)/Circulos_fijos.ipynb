```python
import plotly.express as px
import pandas as pd

# Crear datos de ejemplo para el primer conjunto de puntos
df1 = pd.DataFrame({
    'Tiempo': [1, 2, 3, 4, 5],
    'x': [1, 2, 3, 4, 5],
    'y': [10, 11, 12, 13, 14]
})

# Crear datos de ejemplo para el segundo conjunto de puntos
df2 = pd.DataFrame({
    'Tiempo': [1, 2, 3, 4, 5],
    'x': [5, 4, 3, 2, 1],
    'y': [14, 13, 12, 11, 10]
})

# Concatenar ambos conjuntos de datos
df = pd.concat([df1, df2], keys=['Conjunto1', 'Conjunto2'])

# Crear la gráfica animada
fig = px.scatter(df, x='x', y='y', animation_frame='Tiempo', color=df.index.get_level_values(0),
                 range_x=[0, 6], range_y=[9, 15])

# Personalizar la animación (opcional)
fig.update_layout(
    title='Gráfica Animada con Plotly',
    xaxis_title='Eje X',
    yaxis_title='Eje Y',
)

# Mostrar la gráfica
fig.show()
```

```python
import plotly.express as px
import pandas as pd

# Crear datos de ejemplo para el primer conjunto de puntos
df1 = pd.DataFrame({
    'Tiempo': [1, 2, 3, 4, 5],
    'y': [1] * 5,  # Mantener 'y' constante para el Conjunto1
    'x': [10, 11, 12, 13, 14],
})

# Crear datos de ejemplo para el segundo conjunto de puntos
df2 = pd.DataFrame({
    'Tiempo': [1, 2, 3, 4, 5],
    'y': [2, 1.5, 1, 0.5, 0],  # Ajustar 'y' para que baje a mitad de camino
    'x': [10, 11, 12, 13, 14],
})

# Concatenar ambos conjuntos de datos
df = pd.concat([df1, df2], keys=['Conjunto1', 'Conjunto2'])

# Crear la gráfica animada con camino recorrido
fig = px.scatter(df, x='x', y='y', animation_frame='Tiempo', color=df.index.get_level_values(0),
                 range_x=[9, 15], range_y=[0, 3])

# Personalizar la animación (opcional)
fig.update_layout(
    title='Gráfica Animada con Plotly',
    xaxis_title='Eje X',
    yaxis_title='Eje Y',
    template='plotly_dark',  # Cambiar al tema oscuro
    xaxis=dict(showgrid=False),  # Quitar la cuadrícula del eje X
    yaxis=dict(showgrid=False),  # Quitar la cuadrícula del eje Y
)

# Crear una gráfica adicional para visualizar las trayectorias del recorrido total
fig_total = px.line(df, x='x', y='y', color=df.index.get_level_values(0), range_x=[9, 15], range_y=[0, 3])

# Personalizar la gráfica total (opcional)
fig_total.update_layout(
    title='Trayectorias del Recorrido Total',
    xaxis_title='Eje X',
    yaxis_title='Eje Y',
    template='plotly_dark',  # Cambiar al tema oscuro
    xaxis=dict(showgrid=False),  # Quitar la cuadrícula del eje X
    yaxis=dict(showgrid=False),  # Quitar la cuadrícula del eje Y
)

# Mostrar ambas gráficas
fig.show()
fig_total.show()

```
