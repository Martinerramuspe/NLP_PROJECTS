```python
import pandas as pd

data = {
    'seconds': [0, 4],
     'datos': [3, 2]
}

df = pd.DataFrame(data)
df.index.name = 'ID'

print(df)
```

```python
data1 = {
    'seconds': [0,1,2,3,4],
}

df1 = pd.DataFrame(data1)
df1.index.name = 'ID'

print(df1)
```

```python
def merge_and_add_data(row, df):
    # Obtén el valor de la columna 'seconds' de la fila actual
    seconds_value = row['seconds']

    # Busca el valor correspondiente en la columna 'datos' del DataFrame original (df)
    matching_data = df[df['seconds'] == seconds_value]['datos'].values

    # Devuelve el valor encontrado o NaN si no hay coincidencia
    return matching_data[0] if len(matching_data) > 0 else 0
```

```python
# Aplica la función usando apply y una función lambda
df1['new_column'] = df1.apply(lambda row: merge_and_add_data(row, df), axis=1)
```
